```python
# 每兩個小時抓一次資料


import urllib.request as request
import datetime
from time import time ,sleep
from random import randint
import requests as req
from bs4 import BeautifulSoup as bs
from fake_useragent import UserAgent
from time import time ,sleep
from random import randint
from selenium.webdriver.common.action_chains import ActionChains
import datetime

# 強制等待 (執行期間休息一下)
from time import sleep

# 整理 json 使用的工具
import json

# 執行 command 的時候用的
import os

from time import sleep, time

import schedule

import time
import json

folderPath = "Current_Price"

if not os.path.exists(folderPath):
    os.makedirs(folderPath)


today = datetime.date.today()
today = int(today.strftime('%Y%m%d'))
today = today-19110000
today = str(today)
t1 = today[:3]
t2 = today[3:5]
t3 = today[5:8]



def price(date):
    url = f"https://data.coa.gov.tw/api/v1/AgriProductsTransType/?Start_time={date}&End_time={date}&CropCode=LA1"

    with request.urlopen(url) as rs:
        datas_list = json.load(rs)


    resultes = datas_list['Data']
    
    now = datetime.datetime.now()

    ts = now.strftime('%Y-%m-%d %H:%M:%S')
    
    print('do func time :',ts)    

    file_name = now.strftime('%Y-%m-%d')
        
    for data in resultes:        
        detail = f"TransDate:{data['TransDate']},Avg_Price:{data['Avg_Price']}"
        print(detail)
        with open(f'{folderPath}/{file_name}.txt','a',encoding="UTF-8") as file:
            Jcontent = {
            'DownloadTime':f'{ts}',
            'TransDate':data['TransDate'],
            "CropName": data["CropName"],
            "Avg_Price":data['Avg_Price']
            }
            json.dump(Jcontent,file,ensure_ascii=False)
            
        
        
                  


schedule.every(10).seconds.do(price,date=f"{t1}.{t2}.{t3}")

while True:

    schedule.run_pending()

    time.sleep(1)
```
